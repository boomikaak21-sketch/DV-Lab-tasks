```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------- Load & Clean ----------
file_path = "/content/employee_database (1).csv"
df = pd.read_csv(file_path)

# Keep only needed columns and ensure numeric Salary
df = df[['Department', 'Salary']].copy()
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df = df.dropna(subset=['Department', 'Salary'])

# Filter out very small groups (optional)
group_counts = df['Department'].value_counts()
valid_cats = group_counts[group_counts >= 5].index.tolist()
df = df[df['Department'].isin(valid_cats)]

# Sort categories by median Salary
order = (
    df.groupby('Department')['Salary']
      .median()
      .sort_values(ascending=True)
      .index
      .tolist()
)

# ---------- Simple Gaussian KDE ----------
def gaussian_kde_1d(x_grid, samples):
    samples = np.asarray(samples, dtype=float)
    n = samples.size
    if n == 0:
        return np.zeros_like(x_grid)
    std = samples.std(ddof=1) if n > 1 else 0.0
    if std == 0.0:
        h = max(1e-6, 0.01 * (samples.mean() if samples.mean() != 0 else 1.0))
    else:
        h = 1.06 * std * (n ** (-1/5))
        if h <= 0:
            h = max(1e-6, 0.01 * std)
    u = (x_grid[:, None] - samples[None, :]) / h
    dens = np.exp(-0.5 * u**2) / np.sqrt(2*np.pi)
    dens = dens.sum(axis=1) / (n * h)
    return dens

# Global x-range for Salary
xmin = df['Salary'].min()
xmax = df['Salary'].max()
pad = 0.02 * (xmax - xmin) if xmax > xmin else 1.0
x_grid = np.linspace(xmin - pad, xmax + pad, 512)

# ---------- Ridgeline Plot ----------
fig, ax = plt.subplots(figsize=(10, 6))
offset_step = 1.2
yticks, ytick_labels = [], []

for i, cat in enumerate(order):
    data_cat = df.loc[df['Department'] == cat, 'Salary'].values
    dens = gaussian_kde_1d(x_grid, data_cat)
    if dens.max() > 0:
        dens = dens / dens.max()

    offset = i * offset_step
    ax.fill_between(x_grid, offset, offset + dens, alpha=0.6, linewidth=0.5)
    ax.plot(x_grid, offset + dens, linewidth=0.8)

    yticks.append(offset + 0.5)
    ytick_labels.append(cat)

ax.set_xlabel("Salary")
ax.set_yticks(yticks)
ax.set_yticklabels(ytick_labels)
ax.set_title("Ridgeline Plot (Salary by Department)")
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean dataset
file_path = "/content/employee_database (1).csv"
df = pd.read_csv(file_path)

# Drop rows with missing values in relevant columns
df = df.dropna(subset=['Department', 'Salary'])

# Beeswarm (Swarm) Plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Department', y='Salary', data=df, size=4)
plt.title('Beeswarm Plot: Salary by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
