```
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv('/content/Employee.csv')
# Step 2: Display the original data
print("Original Data:")
print(df.head())

# Step 3: Check for missing values
print("\nMissing values before cleaning:")
print(df.isnull().sum())

# Step 4: Drop rows with missing values
df.dropna(inplace=True)

# Step 5: Filter data for a specific City (e.g., Bangalore)
df_bangalore = df[df['City'] == 'Bangalore'].copy()

# Step 6: Create a new column: Score = Age Ã— ExperienceInCurrentDomain
df_bangalore['Score'] = df_bangalore['Age'] * df_bangalore['ExperienceInCurrentDomain']

# Step 7: View the cleaned and prepared dataset
print("\nPrepared Data (Bangalore City):")
print(df_bangalore.head())

```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv("Employee.csv")

# Step 2: Create a new column - ExperienceScore
df['ExperienceScore'] = df['Age'] * df['ExperienceInCurrentDomain']

# Step 3: Filter for Bangalore city
df_bangalore = df[df['City'] == 'Bangalore']

# Step 4: Set the plot style
sns.set(style="whitegrid")

# Step 5: Plot ExperienceScore over JoiningYear for Bangalore
plt.figure(figsize=(10, 6))
sns.barplot(data=df_bangalore, x='JoiningYear', y='ExperienceScore', palette='Blues_d')

# Step 6: Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Step 7: Add labels and title
plt.title('Experience Score by Joining Year - Bangalore')
plt.xlabel('Joining Year')
plt.ylabel('Experience Score')
plt.tight_layout()

# Step 8: Show the plot
plt.show()
```

```
import pandas as pd
import plotly.express as px

# Step 1: Load dataset
df = pd.read_csv("Employee.csv")

# Step 2: Create a new column - ExperienceScore
df['ExperienceScore'] = df['Age'] * df['ExperienceInCurrentDomain']

# Step 3: Filter for Bangalore city (instead of South region)
df_bangalore = df[df['City'] == 'Bangalore']

# Step 4: Create an interactive line chart
fig = px.line(
    df_bangalore,
    x='JoiningYear',
    y='ExperienceScore',
    title='Experience Score Over Years - Bangalore',
    markers=True,
    labels={'JoiningYear': 'Joining Year', 'ExperienceScore': 'Experience Score'}
)

# Step 5: Show interactive chart
fig.show()
```

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
