```
# Run this whole cell in Google Colab

# 1) Install required packages (quiet)
!pip install wordcloud nltk matplotlib --quiet

# 2) Imports
import re
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 3) Download only stopwords (safe)
nltk.download('stopwords', quiet=True)

# 4) Example text data (replace with your own text if desired)
text_data = """
Technology continues to transform the way humans live, work, and communicate.
Artificial intelligence, automation, and innovation drive smarter realities and
connect data across the internet of things. Cloud computing, machine learning,
and intelligent systems ensure the availability of insights that shape the future.
Reality is being transformed by data-driven approaches that drive sustainability,
scalability, and smarter infrastructure for humans globally.
"""

# 5) Preprocessing: extract alphabetic words and remove stopwords
stop_words = set(stopwords.words('english'))
tokens = re.findall(r"\b[a-zA-Z]+\b", text_data.lower())          # safe tokenization
filtered_words = [w for w in tokens if w not in stop_words and len(w) > 1]

# 6) Join back to string for WordCloud
clean_text = " ".join(filtered_words)

# 7) Generate WordCloud
wc = WordCloud(
    width=1000,
    height=600,
    background_color='white',
    colormap='viridis',
    max_words=200,
    contour_width=0
).generate(clean_text)

# 8) Display
plt.figure(figsize=(12,8))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud Visualization", fontsize=18)
plt.show()
```
