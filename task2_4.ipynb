```python
x import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Education.csv")

# Set style
sns.set(style="whitegrid")

# ===============================
#  CATEGORICAL DATA VISUALIZATION
# ===============================

# Bar Chart - Frequency of each Field of Study
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Field_of_Study', palette='pastel')
plt.title("Bar Chart: Frequency of Fields of Study")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Education.csv")

# Set style
sns.set(style="whitegrid")

# Pie Chart - Distribution of Fields of Study
field_counts = df['Field_of_Study'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(field_counts, labels=field_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart: Field of Study Distribution")
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Education.csv")

# Set style
sns.set(style="whitegrid")

# Histogram - University GPA Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['University_GPA'], kde=False, color='skyblue', bins=15)
plt.title("Histogram: University GPA Distribution")
plt.xlabel("University GPA")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Education.csv")

# Set style
sns.set(style="whitegrid")

# Density Plot - Starting Salary Distribution
plt.figure(figsize=(8, 6))
sns.kdeplot(df['Starting_Salary'], fill=True, color='orange')
plt.title("Density Plot: Starting Salary Distribution")
plt.xlabel("Starting Salary ($)")
plt.ylabel("Density")
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Education.csv")

# Set style
sns.set(style="whitegrid")

# Rug Plot - SAT Score Distribution
plt.figure(figsize=(10, 2))
sns.rugplot(df['SAT_Score'], height=0.05, color='blue')
plt.title("Rug Plot: SAT Score Distribution")
plt.xlabel("SAT Score")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Education.csv")

# Set style
sns.set(style="whitegrid")

# Strip Plot - Starting Salary by Field of Study
plt.figure(figsize=(12, 8))
sns.stripplot(x='Starting_Salary', y='Field_of_Study', data=df, color='green', alpha=0.7)
plt.title("Strip Plot: Starting Salary Distribution by Field of Study")
plt.xlabel("Starting Salary ($)")
plt.ylabel("Field of Study")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Education.csv")

# Set style
sns.set(style="whitegrid")

# Swarm Plot - Starting Salary by Field of Study
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Starting_Salary', y='Field_of_Study', data=df, size=3)
plt.title("Swarm Plot: Starting Salary Distribution by Field of Study")
plt.xlabel("Starting Salary ($)")
plt.ylabel("Field of Study")
plt.show()
```
