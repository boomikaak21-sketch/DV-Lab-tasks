```
# ---- Step 1: Import Libraries ----
import networkx as nx
import matplotlib.pyplot as plt

# ---- Step 2: Create a Larger Sample Dataset (Edges) ----
edges = [
    ("Alice", "Bob"), ("Alice", "Cathy"), ("Alice", "David"),
    ("Bob", "Eve"), ("Cathy", "Frank"), ("David", "Grace"),
    ("Eve", "Helen"), ("Frank", "Grace"), ("Helen", "Ivy"),
    ("Grace", "Jack"), ("Ivy", "Jack"), ("Jack", "Kate"),
    ("Kate", "Leo"), ("Leo", "Mona"), ("Mona", "Nina"),
    ("Nina", "Oscar"), ("Oscar", "Paul"), ("Paul", "Quinn"),
    ("Quinn", "Rita"), ("Rita", "Sam"), ("Sam", "Tina"),
    ("Tina", "Uma"), ("Uma", "Victor"), ("Victor", "Wendy"),
    ("Wendy", "Xavier"), ("Xavier", "Yara"), ("Yara", "Zack"),
    ("Zack", "Alice"), ("Mona", "Eve"), ("Frank", "Oscar"),
    ("Helen", "Leo"), ("David", "Nina"), ("Grace", "Rita"),
    ("Jack", "Tina"), ("Kate", "Uma"), ("Leo", "Victor"),
    ("Mona", "Wendy"), ("Nina", "Xavier"), ("Oscar", "Yara"),
    ("Paul", "Zack")
]

# ---- Step 3: Create the Graph ----
G = nx.Graph()
G.add_edges_from(edges)

# ---- Step 4: Compute Force-based Layout ----
pos = nx.spring_layout(G, seed=42, k=0.5, iterations=100)

# ---- Step 5: Visualize the Graph ----
plt.figure(figsize=(12, 10))
nx.draw_networkx_nodes(G, pos, node_size=1000, node_color='skyblue', edgecolors='black')
nx.draw_networkx_edges(G, pos, width=2, alpha=0.4, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=9, font_weight='bold', font_color='darkblue')

plt.title("Enhanced Force-based Layout Visualization", fontsize=16, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()
```
